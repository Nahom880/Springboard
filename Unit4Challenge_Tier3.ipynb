```python
# Limit the amount of temporal data-points by extracting the year from every value in Month column

# 1. Look through the `Month` column using lambda function 
# 2. Extract the year from each individual value in that column 
# 3. Store that corresponding year as separate column. Use a .loc function to avoid Chained Indexing  

df.loc[:, 'Year'] = df['Month'].apply(lambda t: t.year)
```

```python
# Calling the tail() method on df to see the Column 'Year' addition
df.tail()
```

```python
# Using the function 'groupby' to calculate the mean for each year and Borough. 
## The variables Borough and Year are now indices 
dfg = df.groupby(by=['London_Borough', 'Year'])['Average_price'].mean().reset_index()
dfg.sample(10)
```

```python
# Call the head() method on the new DataFrame dfg
dfg.head()
```

```python
# 3. Modelling 

# Create a function that will calculate a ratio of house prices, that compares the price of a house in 2018 to the price in 1998.

# These are the steps: 
# 1. Filter dfg, specifically where this filter constrains the London_Borough, as an argument
# 2. Get the Average Price for that borough for 1998 and, seperately, for 2018. 
# 3. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018. 
# 4. Return that ratio. 

# Use the function to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018

# Heads up: By using .iloc[0], explicitly selecting the first element of the Series, which avoids the FutureWarning

# Here is the function 

def create_price_ratio(d):
    y1998 = float(d['Average_price'][d['Year']==1998].iloc[0])
    y2018 = float(d['Average_price'][d['Year']==2018].iloc[0])
    ratio = [y2018/y1998]
    return ratio
```

```python
#  Test the function by calling it with the following argument:

create_price_ratio(dfg[dfg['London_Borough']=='Barking & Dagenham'])
```

```python
# Ultimately use the create_price_ratio function to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

# First, make an empty dictionary, called final, where we'll store our ratios for each unique London_Borough.
final = {}
```

```python
# Declare a for loop that will iterate through each of the unique elements of the 'London_Borough' column of the DataFrame dfg.

# Call the iterator variable 'b'. 
for b in dfg['London_Borough'].unique():
    # Make the parameter to our create_price_ratio function
    borough = dfg[dfg['London_Borough'] == b]
    # Make a new entry in the final dictionary whose value's the result of calling create_price_ratio with the argument: borough
    final[b] = create_price_ratio(borough)
# Use the function and incorporate that into a new key of the dictionary 
print(final) 
```

```python
# Make a variable called df_ratios, and assign it the result of calling the DataFrame method on the dictionary final
df_ratios = pd.DataFrame(final)
```

```python
# Call the head() method on this variable to check it out. 
df_ratios.head()
```

```python
# Transpose it and reset the index! 
df_ratios_T = df_ratios.T
df_ratios = df_ratios_T.reset_index()
df_ratios.head()
```

```python
# Rename the 'index' column as 'London_Borough', and the '0' column to '2018'.
df_ratios.rename(columns={'index':'Borough', 0:'2018'}, inplace=True)
df_ratios.head()
```

```python
# Create a variable called top15 to assign the result of sort_values() in descending order and select the top 15 boroughs on df_ratios.

top15 = df_ratios.sort_values(by='2018',ascending=False).head(15)
print(top15)
```

```python
# Plot the boroughs that have seen the greatest changes in price.

ax = top15[['Borough','2018']].plot(kind='bar')

ax.set_xticklabels(top15.Borough)
```
